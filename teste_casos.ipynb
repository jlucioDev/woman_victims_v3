```python
import pandas as pd

# Carregando o dataset com os municípios classificados
df_munic = pd.read_csv('datasets/df_classified_pa_casos.csv')
# coluna_uf = 'vitima_uf'
# coluna_municipio ='vitima_municipio'
# coluna_uf = 'uf'
# coluna_municipio ='municipio'
coluna_uf = 'UF'
coluna_municipio ='Município'

pasta = 'C:/Users/JrDev/Downloads/ligue180-'
ano = '2023'
parte ='2'
nome_coluna = f'qty_casos_{ano}'
df = pd.read_csv(f'{pasta}{ano}-parte{parte}.csv', sep=';')
#df = pd.read_csv(f'{pasta}{ano}-parte{parte}.csv', sep=';')

df.head(3)
```

```python
df_munic
```

```python

df[coluna_uf] = df[coluna_uf].str.strip()
#df[coluna_municipio] = df[coluna_municipio].str.lower()
df[coluna_municipio]  = df[coluna_municipio].str.title()
#df[coluna_municipio] = df[coluna_municipio].apply(lambda x: x.split('(')[0].strip())
df[coluna_municipio] = df[coluna_municipio].str.slice(9).str.strip()

df_filtrado = df.query(f"{coluna_uf} == 'PA'")

df_filtrado.head(3)
```

```python
# Agrupe os dados por 'municipio' e conte o número de ocorrências em cada grupo
contagem_casos_por_municipio = df_filtrado.groupby(coluna_municipio).size().reset_index(name=nome_coluna)
contagem_casos_por_municipio[nome_coluna] = contagem_casos_por_municipio[nome_coluna].astype(float)
# Exiba a contagem de casos por município
contagem_casos_por_municipio
```

```python

# Use uma função lambda com str.split() para dividir o valor da coluna pelo parêntese e pegar o primeiro elemento

contagem_casos_por_municipio
```

```python
# Verifique as colunas que contêm '2018' em seus nomes
# Se houver colunas com 'ano', some os valores dessas colunas
if nome_coluna in df_munic.columns:
    nome_coluna_2 = f'{nome_coluna}_'
    contagem_casos_por_municipio.rename(columns={nome_coluna: nome_coluna_2}, inplace=True)

    # Realize um merge entre os DataFrames com base na coluna 'localidade' em df_munic e 'municipio' em contagem_casos_por_municipio
    df_merged = df_munic.merge(contagem_casos_por_municipio, left_on='localidade', right_on=coluna_municipio, how='left')

    # Preencha os valores ausentes (caso o município não exista na contagem de casos) com zero
    df_merged[nome_coluna_2] = df_merged[nome_coluna_2].fillna(0)

    # Remova a coluna 'municipio' resultante do merge
    df_merged.drop(coluna_municipio, axis=1, inplace=True)
    df_merged.drop('Unnamed: 0', axis=1, inplace=True)
    
    df_merged['somado'] = df_merged[nome_coluna] + df_merged[nome_coluna_2]
    df_merged.drop(nome_coluna, axis=1, inplace=True)
    df_merged.drop(nome_coluna_2, axis=1, inplace=True)
    # Use o método rename() para renomear a coluna
    df_merged.rename(columns={'somado': nome_coluna}, inplace=True)

    print('Somado')

    
else:
    
    # Realize um merge entre os DataFrames com base na coluna 'localidade' em df_munic e 'municipio' em contagem_casos_por_municipio
    df_merged = df_munic.merge(contagem_casos_por_municipio, left_on='localidade', right_on=coluna_municipio, how='left')

    # Preencha os valores ausentes (caso o município não exista na contagem de casos) com zero
    df_merged[nome_coluna] = df_merged[nome_coluna].fillna(0)

    # Remova a coluna 'municipio' resultante do merge
    df_merged.drop(coluna_municipio, axis=1, inplace=True)
    df_merged.drop('Unnamed: 0', axis=1, inplace=True)
    
    print('Agregado')
# Exiba o DataFrame resultante df_merged com a nova coluna 'qtd_casos'
df_merged
```

```python
df_merged.to_csv('datasets/df_classified_pa_casos.csv')
```
